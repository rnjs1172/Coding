```python
from urllib.parse import urljoin 

def cafe_joonggo(page):
    url_main = "https://m.cafe.naver.com"
    url_host =  "/ArticleAllListAjax.nhn"
    data = []
    for num in range(page):
        params = {
            'search.clubid':'10050146',
            'search.boardtype':'L',
            'search.menuid': '1256',
            'search.questionTab':'A',
            'search.totalCount':'201',
            'search.page': num
        }
        list_url = urljoin(url_main,url_host)
        article_html = requests.get(list_url,params=params).text
        article_soup = BeautifulSoup(article_html,"html.parser")
        
        for tag in article_soup.select(".list_area .board_box"):
            artid_title = tag.find(class_ = "tit").text.strip()
            artid_url = tag.find("a")["href"]
            sample = []
            sample = "".join(artid_title + artid_url)
            data.append(sample)
    return(data)
```

### 10.2  MechanicalSoup 라이브러리 이용해 보자!

```python
import mechanicalsoup

## 일단 로그인 창의 url을 
log_url = "https://nid.naver.com/nidlogin.login?"### 뒤에 있는 get 인자는 필요 없다.

### mecha 를 쓸떄 필요한 url을 담을 변수를 만들자!

browser_1 = mechanicalsoup.StatefulBrowser()
browser_1.open(log_url) ### 이 명령어에 내부적으로 requests.get 함수가 쓰인다.
```
